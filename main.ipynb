```python
from main import (
    load_dataset,
    general_describe,
    generate_vis,
    generate_dist,
    visualize_boxplot,
)

dataset = "./StudentPerformanceFactors.csv"

general_df = load_dataset(dataset)

assert general_df is not None
assert general_df.shape == (6607, 20)
general_df.head()
```

```python
column_to_analyze = "Hours_Studied"
summary_stats = general_describe(general_df, column_to_analyze)

describe_df = general_df.describe().loc[:, column_to_analyze]

# Assertions to check the statistics
assert round(summary_stats["mean"], 2) == round(
    describe_df["mean"], 2
), "Mean is incorrect"
assert round(summary_stats["std"], 2) == round(
    describe_df["std"], 2
), "Standard deviation is incorrect"
assert round(summary_stats["50%"], 2) == round(
    describe_df["50%"], 2
), "Median (50%) is incorrect"
assert round(summary_stats["25%"], 2) == round(
    describe_df["25%"], 2
), "25th percentile is incorrect"
assert round(summary_stats["75%"], 2) == round(
    describe_df["75%"], 2
), "75th percentile is incorrect"

summary_stats
```
